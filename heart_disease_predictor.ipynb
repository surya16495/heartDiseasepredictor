```python
import numpy as np 
import pandas as pd 

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression 
```

```python
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("heart.csv",encoding='latin-1')
```

- 1 $\rightarrow$ Heart Disease

#### Splitting Features and Target

```python
X=df.drop(columns='target',axis=1)
X
```

```python
Y=df['target']
Y
```

#### Training and Testing

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

#### Model Building 

LogisticRegression is used to create the model where end of the day it is a binary classification

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

#### Evaluation



```python
# acuuracy on training data
X_train_prediction=model.predict(X_train)
training_accuracy=accuracy_score(X_train_prediction,Y_train)
print(f"The training acccuracy is{training_accuracy}")
```

```python
# acuuracy on testing data
X_test_prediction=model.predict(X_test)
testing_accuracy=accuracy_score(X_test_prediction,Y_test)
print(f"The testing acccuracy is{testing_accuracy}")
```

### Building the Predicting system:

```python
# Tuple of the feature data which is my input
input_data=(61,1,2,150,243,1,1,137,1,1,1,0,2)
#converting it into anumpy array
input_data=np.asarray(input_data)
#reshaping into required input form
input_data=input_data.reshape(1,-1)
#with the input I am predicting
prediction=model.predict(input_data)
prediction

if prediction[0]==0:
    print("Good News the patient does'nt have any heart disease")
else:
    print("The Patient should visit the doctor")
```
